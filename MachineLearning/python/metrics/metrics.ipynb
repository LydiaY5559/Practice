## F1 Score

下图横/纵轴分别为precision和recall

在P/R对应的值下，F1 score的热力分布

可以看到两个都大时，F1也会很大；两个都小时，F1会很小

另外，可以看到P=0.4, R=0.3跟P=1.0, R=0.2差距很小（所以某种程度上来说，提升较小的那个可能比较划算）

```python
l = 10
f1_array = np.zeros((l, l))
f1_score = lambda p, r: 2*p*r/(p+r) if p*r > 0 else 0
for i in range(l):
    for j in range(l):
        f1_array[i][j] = f1_score((i+1)/l, (j+1)/l)
f1_df = pd.DataFrame(f1_array, index=np.arange(1,l+1)/l, columns=np.arange(1,l+1)/l)
sns.heatmap(f1_df, annot=True)
```
