```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def sinplot(filp=1):
    x = np.linspace(0, 14, 100)
    for i in range(0,7):
        plt.plot(x, np.sin(x+0.5*i)*(7-i)*filp)
```

```python
# --------------------------------------------------------------------------------
# 控制展示样式
# --------------------------------------------------------------------------------
```

```python
sinplot()
```

```python
import seaborn as sns
sinplot()
```

```python
# set_style(): darkgrid, whitegrid, dark, white, and ticks
sns.set_style("ticks")
sinplot()
```

```python
sinplot()
sns.despine() # 去掉ticks上方和右方的脊
```

```python
data = np.random.normal(size=(20, 6)) + np.arange(6) / 2
sns.violinplot(data=data)
sns.despine(offset=10, trim=True) # x y轴标的距离
```

```python
# 箱线图
sns.set_style("whitegrid")
sns.boxplot(data=data)
sns.despine(left=True)
```

```python
# 临时修改style
with sns.axes_style("darkgrid"):
    plt.subplot(211)
    sinplot()
plt.subplot(212)
sinplot(-1)
```

```python
# 获取axes的style属性
sns.axes_style()
```

```python
# 通过dict修改style
sns.set_style("darkgrid", {"axes.facecolor": ".9"})
sinplot()
```

```python
# 重设所有设置
sns.set()
sinplot()
```

```python
# paper, notebook, talk, and poster
sns.set_context("poster")
sinplot()
```

```python
sns.set_context("notebook", font_scale=1.5, rc={"lines.linewidth": 2.5})
sinplot()
```

```python
# default color cycle
current_palette = sns.color_palette()
sns.palplot(current_palette)
```

```python
# more color, using 'hls'
sns.palplot( sns.color_palette('hls', 8) ) # 8 is color num
# sns.palplot( sns.hls_palette(8) ) # same
```

```python
# control lightness and saturation of the color
sns.palplot( sns.hls_palette(8, l=.3, s=.8) )
```
